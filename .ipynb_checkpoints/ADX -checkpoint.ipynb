```python
from tvDatafeed import TvDatafeed, Interval
import pandas as pd
import pandas_ta as ta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os

# Autenticao no TvDatafeed
# Substitua 'your_username' e 'your_password' com suas credenciais do TradingView
username = os.getenv('TRADINGVIEW_USERNAME')
password = os.getenv('TRADINGVIEW_PASSWORD')

username
password
tv = TvDatafeed(username, password)

ticker = 'BPAC11'

# Obtm os dados de preo das aes com intervalo de 1 hora e 1000 barras
data = tv.get_hist(symbol=ticker, exchange='BMFBOVESPA', interval=Interval.in_1_hour, n_bars=1000)

# Calcula o ADX usando pandas_ta (isso tambm incluir +DI e -DI no DataFrame)
data.ta.adx(append=True)

# Cria subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plotagem do preo de fechamento
ax1.plot(data.index, data['close'], label='Preo de Fechamento', color='blue')
ax1.set_ylabel('Preo de Fechamento')
ax1.legend(loc='upper left')
ax1.set_title(f'Preo de Fechamento e Indicadores ADX/DI para {ticker}')

# Formatao do eixo de datas para o subplot superior
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)

# Plotagem do ADX, DI+ e DI- no subplot inferior
ax2.plot(data.index, data['ADX_14'], label='ADX', color='black')
ax2.plot(data.index, data['DMP_14'], label='DI+', color='green')
ax2.plot(data.index, data['DMN_14'], label='DI-', color='red')
ax2.set_ylabel('ADX / DI+ / DI-')
ax2.axhline(25, color='grey', linestyle='--', label='Nvel Significativo (25)')
ax2.legend(loc='upper left')

# Formatao do eixo de datas para o subplot inferior
ax2.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
from tvDatafeed import TvDatafeed, Interval
import pandas as pd
import pandas_ta as ta
import mplfinance as mpf

username = os.getenv('TRADINGVIEW_USERNAME')
password = os.getenv('TRADINGVIEW_PASSWORD')

username
password
tv = TvDatafeed(username, password)

ticker = 'pomo3'

# Obtm os dados de preo das aes com intervalo dirio e 1000 barras
data = tv.get_hist(symbol=ticker, exchange='BMFBOVESPA', interval=Interval.in_daily, n_bars=500)

# Calcula o ADX usando pandas_ta (isso tambm incluir +DI e -DI no DataFrame)
data.ta.adx(append=True)

# Ajusta o ndice para ser do tipo datetime, necessrio para mplfinance
data.index = pd.to_datetime(data.index)

# Configura o estilo dos grficos mplfinance
mpf_style = mpf.make_mpf_style(base_mpl_style='seaborn', rc={'font.size': 8})

# Plota o grfico de candlestick separadamente
mpf.plot(data, type='candle', style=mpf_style, volume=False, figratio=(12, 4), title=f'{ticker} Candlestick')

# Cria um novo grfico para os indicadores tcnicos usando matplotlib
fig, ax = plt.subplots(figsize=(12, 4))

# Plota o ADX, DI+ e DI- no mesmo grfico
ax.plot(data.index, data['ADX_14'], label='ADX', color='black', linewidth=1)
ax.plot(data.index, data['DMP_14'], label='DI+', color='green', linewidth=1)
ax.plot(data.index, data['DMN_14'], label='DI-', color='red', linewidth=1)

# Adiciona uma linha horizontal para indicar o nvel significativo do ADX
ax.axhline(25, color='grey', linestyle='--', label='Nvel Significativo (25)')

# Configuraes de legenda e ttulo
ax.legend(loc='upper left')
ax.set_title(f'{ticker} ADX/DI Indicators')
ax.set_ylabel('ADX / DI+ / DI-')

# Melhora a formatao do eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Calcula a diferena entre DI+ e DI- e sua diferena absoluta
soma_di = data['DMP_14'].iloc[-1] - data['DMN_14'].iloc[-1]
dif_abs_di = abs(soma_di)

# Determina a fora da tendncia
forca_tendencia = "forte" if data['ADX_14'].iloc[-1] > 25 else "fraca"

# Determina a direo da tendncia
direcao_tendencia = "de alta" if soma_di > 0 else "de baixa"

# Verifica se o mercado est lateralizado com base na diferena absoluta entre DI+ e DI-
if dif_abs_di <= 5:
    estado_mercado = "lateralizado"
else:
    estado_mercado = "em tendncia"

# Combina a direo e a fora da tendncia com o estado do mercado
descricao_tendencia = f"{direcao_tendencia} {forca_tendencia}" if estado_mercado == "em tendncia" else "lateralizado"

# Exibe os valores de ADX e a fora da tendncia
print(f"Valores finais ")
print(f"ADX: {data['ADX_14'].iloc[-1]:.2f} (Tendncia {forca_tendencia})")

# Exibe os valores de DI+ e DI-
print(f"DI+: {data['DMP_14'].iloc[-1]:.2f}")
print(f"DI-: {data['DMN_14'].iloc[-1]:.2f}")

# Exibe a diferena absoluta entre DI+ e DI- e o estado do mercado
print(f"Diferena absoluta DI+ - DI-: {dif_abs_di:.2f} ({estado_mercado})")

# Exibe a descrio da tendncia
print(f"Tendncia: {descricao_tendencia}\n")

```

```python
from tvDatafeed import TvDatafeed, Interval
import pandas as pd
import pandas_ta as ta

# Autenticao no TvDatafeed
username
password
tv = TvDatafeed(username, password)

# L a lista de tickers do arquivo Excel
tickers_df = pd.read_excel('/Users/flaviomotta/PROJETOS ANALISE FINANCEIRA/DADOSDEMERCADO.COM.BR/tickers_filtrados_vol.xlsx', engine='openpyxl')
tickers = tickers_df['Ticker']

# Lista para armazenar os DataFrames de resultados temporrios
resultados_temp = []

# Itera sobre cada ticker e busca os dados, calcula o ADX e a diferena entre DI+ e DI-
for ticker in tickers:
    try:
        # Obtm os dados de preo das aes com intervalo dirio e 500 barras
        data = tv.get_hist(symbol=ticker, exchange='BMFBOVESPA', interval=Interval.in_daily, n_bars=500)
        # Calcula o ADX usando pandas_ta
        data.ta.adx(append=True)
        # Calcula a diferena entre DI+ e DI-
        data['DI_diff'] = abs(data['DMP_14'] - data['DMN_14'])
        # Adiciona os resultados ao DataFrame temporrio
        ultima_linha = data.iloc[-1]
        resultados_temp.append(pd.DataFrame({'Ticker': [ticker], 'ADX': [ultima_linha['ADX_14']], 'DI+ - DI-': [ultima_linha['DI_diff']]}))
    except Exception as e:
        print(f"Erro ao processar {ticker}: {e}")

# Concatena todos os DataFrames temporrios em um DataFrame final
resultados = pd.concat(resultados_temp)

# Ordena os resultados pelo ADX de forma decrescente
resultados.sort_values(by='ADX', ascending=False, inplace=True)

# Configura o Pandas para mostrar todas as linhas
pd.set_option('display.max_rows', None)

# Exibe o DataFrame como uma tabela HTML
from IPython.display import display, HTML
display(HTML(resultados.to_html(index=False)))
```
