```python
from tvDatafeed import TvDatafeed, Interval
import pandas as pd
import matplotlib.pyplot as plt
import os 

# Substitua com suas credenciais
username = os.getenv('TRADINGVIEW_USERNAME')
password = os.getenv('TRADINGVIEW_PASSWORD')


tv = TvDatafeed(username, password)

def get_normalized_data(ticker, start_date):
    # Baixar os dados no timeframe de 1 hora
    data = tv.get_hist(symbol=ticker, exchange="BMFBOVESPA",
                       interval=Interval.in_daily, n_bars=10000, extended_session=False)
    
    # Filtrar dados para o período após a data de início
    data = data[data.index >= pd.to_datetime(start_date)]

    # Normalizar os dados de fechamento
    normalized_data = (data['close'] / data['close'].iloc[0]) * 100
    return data, normalized_data

# Lista de ativos do varejo
tickers = ['ASAI3', 'CEAB3', 'MGLU3', 'BHIA3', 'GUAR3', 'LREN3', 'LJQQ3']

# Data inicial para análise
start_date = '2024-01-01'  # Altere esta data conforme necessário

# Criar um gráfico e um DataFrame para as estatísticas
plt.figure(figsize=(10, 6))
stats = []

for ticker in tickers:
    data, normalized_data = get_normalized_data(ticker, start_date)
    plt.plot(normalized_data, label=ticker)

    # Adiciona as estatísticas ao DataFrame
    initial_price = data['close'].iloc[0]
    final_price = data['close'].iloc[-1]
    variation_percent = (final_price - initial_price) / initial_price * 100
    stats.append([ticker, initial_price, final_price, variation_percent])

plt.title('Comparação da Variação Percentual de Ativos do Varejo desde ' + start_date)
plt.xlabel('Data')
plt.ylabel('Variação Percentual (Normalizada)')
plt.legend()
plt.show()

# Criando o DataFrame com as estatísticas
stats_df = pd.DataFrame(stats, columns=['Ticker', 'Preço Inicial', 'Preço Final', 'Variação Percentual'])
# Ordenando o DataFrame por variação percentual de forma decrescente
sorted_stats_df = stats_df.sort_values(by='Variação Percentual', ascending=False)

# Exibindo o DataFrame ordenado
print(sorted_stats_df)
```
